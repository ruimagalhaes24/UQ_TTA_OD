```python
import numpy as np
from matplotlib import pyplot as plt

plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

def entropy_multivariate_normal(x):
   return 0.5 * np.log(2*np.pi*np.exp(1)*x)

x = np.linspace(1e-16, 1000, 100)

plt.plot(x, entropy_multivariate_normal(x), color='red')

plt.show()
```
